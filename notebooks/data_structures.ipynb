```python
### File Handling

# "r"- read, error if file doesnot exist
# "w"- write, create file if not exist, overwrite if exist
# "a"- append, create file if not exist, add data at end if exist

# "t"- text mode
# "b"- binary mode

filename= "/home/olive/Desktop/hackademia/sample.txt"
mode= "w"

files= open(filename, mode)
# files.read()
files.write("This is a sample text, success!")
```

```python
first_set = {23, 42, 65, 57, 78, 83, 29}
second_set = {57, 83, 29, 67, 73, 43, 48}

intersection_items= first_set.intersection(second_set)

for items in intersection_items:
    first_set.remove(items)

print("First set after removing common element", first_set)
```

```python
difference_set= first_set.difference(second_set)
print(difference_set)
```

```python
## Nested Dictionaries
animals= {
    "dog": {
        "name": "Buddy",
        "age": 5,
        "breed": "Golden Retriever"
    },
    "cat": {
        "name": "Whiskers",
        "age": 3,
        "breed": "Siamese"
    },
    "mouse": {
        "name": "Jerry",
        "age": 1,
        "breed": "House Mouse"
    }
}   
```

```python
# access items in dictionaries

print(animals.keys())
print(animals.values())
```

```python
animals.get("dog")
```

```python
type(animals.get("dog"))
```

```python
animals.get("dog").keys()
animals.get("dog").values()
```

```python
print(animals["dog"]["name"])  # Output: Buddy
print(animals["cat"]["age"])   # Output: 3
print(animals["mouse"]["breed"])  # Output: House Mouse
```
