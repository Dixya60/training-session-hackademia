```python
# operations on strings

# slicing
b= "Hello, welcome to Hackademia!"

b[0]
```

```python
# b[4:11] # slicing from index 4 to 10

b[5:] # slicing from index 5 to end

b[:5] # slicing from start to index 4

b[::-1]  # reverse the string

b[:] # entire string

b[:10] # first 10 characters
```

```python
b[:-4] # all except last character

b= "hello world"

# slice to get llo using negative indexing
b[-9:-6]


b[:-5] # this will give all except last 5 characters
```

```python
# modifying the strings case

all_caps= b.upper()
print(all_caps)
all_lower= b.lower()
print(all_lower)
```
