```python
## Parameters and Arguments
# A parameter is the variable listed inside the parentheses in the function definition.
# An argument is the value that is sent to the function when it is called.
```

```python
def my_function(fname, lname= "john"):
    print("Hello " + fname + " " + lname)

my_function("johnson", "doe")
```

```python
# positional arguments: when we are calling a function with arguments without using keywords,
#  they are called positional arguments

def my_function(animal, name):
    print("I have a ", animal + " named " + name)

my_function("dog", "Buddy")  # correct
my_function("Buddy", "dog")  # incorrect order
```

```python
# functions should be called with correct number of arguments
# *args, **kwargs

def unknown_function(*args, **kwargs):
    print("Positional arguments:", args)
    print("Keyword arguments:", kwargs)

unknown_function([1, 2, 3,5,6,7], name="Alice", age=30)
```

```python
def my_function(*fruits):
    for fruit in fruits:
        print("I like", fruit)

fruits=["apple", "pomegranate", "banana"]

my_function(fruits)
```

```python
# dictionary unpacking as function arguments
def display_info(name, age):
    print(f"Name: {name}, Age: {age}")

info = {'name': 'Bob', 'age': 25}
display_info(**info)
```

```python
# implement checking for even or odd number
status= ""
def check_even_odd(number):
    if number % 2==0:
        print(f"{number} is even")
        status= "even"
        return status
    else:
        status= "odd"
        print(f"{number} is odd")
        return status
```

```python
check_even_odd(55)
```

```python
status= check_even_odd(44)

print(status)
```

```python
# define a function or converting fahrenheit to celcius
def fahrenheit_to_celcius(temp):
    celcius= (temp -32) *5/9 # formula
    return celcius
```

```python
# call a function
result= fahrenheit_to_celcius(86)
print(result)
```

```python
# parameters vs Arguments

# argument is the actual value that we use while calling the function
# parameter is the variable that we use in the function definition to represent the value that will be passed to the function

def greet(name):  # name is parameter
    print(f"Hello, {name}!")

greet("Alice")  # "Alice" is argument
greet("Bob")    # "Bob" is argument
```

```python
# functions with 2 arguments
def add_numbers(num1, num2):  # num1 and num2 are parameters
    return num1 + num2

sum_result= add_numbers(5, 10)  # 5 and 10 are arguments
print(sum_result)
```

```python
# keyword arguments
def describe_person(name: None, age:int, city:str):
    print(f"{name} is {age} years old and lives in {city}.")
```

```python
describe_person(name= "John", age=30, city="New York")
describe_person(city="Los Angeles", name="Alice", age=25)
```

```python
# positional arguments
def describe_pet(pet_name, pet_type):
    print(f"I have a {pet_type} named {pet_name}.")

describe_pet(["Buddy", "xyz"], ["dog"])
describe_pet("Whiskers", "cat")
```
