#### Defining an array

```python
import numpy as np

arr= np.array([[1,2,3], [4,5,6],[22,33,44], [33,21,11]])

print("-"*20)
print(arr)
print("The shape of an array is : ", arr.shape)
print("The dimension of array: ", arr.ndim)
print("Total items in ana array", arr.size)
print("Item size of array", arr.itemsize)
```
