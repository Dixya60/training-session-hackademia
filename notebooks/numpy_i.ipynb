```python
empty_arr= np.empty(25)
# print(empty_arr)

x= np.array([[1,2,3], [4,5,7],[21,89,43], [33,55,66]])
v= np.array([1,0,1])

# print(x.shape)

y= np.empty_like(x)
# print(y)
# print(y.shape)

for i in range(4):
    # add vector v to each row of the matrix with the same shape as x

    print(x[i, :])
    y[i, :]= x[i, :]+ v

print(y)

# print(x)
# print(x.shape)
# print(v.shape)
print("=="*20)
print(x + v) # broadcasting
```

```python
v= np.array([1,2,3])
w= np.array([4,5])

print(v, v.shape)
print(v.ndim)
print(w, w.shape)
```

```python
# To compute an outer product, we first reshape v to be a column
# vector of shape (3, 1); we can then broadcast it against w to yield
# an output of shape (3, 2), which is the outer product of v and w:
# [[ 4  5]
#  [ 8 10]
#  [12 15]]

# reshape v into column to perform outer product
reshaped= np.reshape(v, (3,1))
print(reshaped)
print(reshaped.shape)
# print(reshaped.ndim)

print(w.shape)
```

```python
print(reshaped* w)
```
