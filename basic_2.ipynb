```python
# multiple conditions scenarios

y = input("Enter a number: ")
y = int(y)

# check for multiple conditions
if y>0:
    print("the number is positive")
elif y<0:
    print("the number is negative")
elif y==0:
    print("the number is zero")
else:
    print("invalid input")
```

```python
# Local and global variables

var1= 10  # global variable
var2= 5  

def my_function():
    print("Inside the function, var1:", var1)
    print("Inside the function, var2:", var2)

my_function()
```

```python
var1= 55
var2= 45

def print_function():
    var1= 100  # local variable
    var2= 200  # local variable
    print("Inside the function, var1:", var1)
    print("Inside the function, var2:", var2)
```

```python
# print_function()

# if we want to access the global variables inside the function even after defining local variables with the same name, we can use the 'global' keyword
def access_global():
    global var1
    global var2
    var1= 300  # modifying global variable
    var2= 400  # modifying global variable

    print("Accessing global var1:", var1)
    print("Accessing global var2:", var2)
```

```python
access_global()
```

```python
### Using loops in python

# for loop
for i in range(5):
    print("Iteration:", i)
```

```python
# range function variations
print("Range with one argument:")
for i in range(0, 20, 2):
    print(i)
```

```python
# using loops with break, continue, pass statements
print("Using break statement:")
for i in range(10):
    if i == 5:
        print("Breaking the loop at i =", i)
        break
        print("This will not be printed")
    print(i)
```

```python
# using continue statement
print("Using continue statement:")
for i in range(10):
    if i%2 ==0:
        print("Skipping even number i =", i)
        continue
    print(i)
```

```python
# pass statement
print("Using pass statement:")
for i in range(5):
    if i == 3:
        print("Passing at i =", i)
        pass  # does nothing, just a placeholder
    print(i)
```

```python
# while loops
i=1
while i<10:
    print("While loop iteration:", i)
    i +=1
    
```
