#### Loops and iterations

```python
# for loops
# while loops
# loop control statements: break, continue, pass

# why do we need loops?
# to iterate over a block of code multiple times until a certain condition is met
```

```python
# print(1)
# print(2)
# print(3)

# another way is to use loops
# for i in range(1, 4):
#     print(i)

# here range function generates a sequence of numbers from 1 to 3 (4 is excluded)
# other parameters of range function are start, stop, step
for i in range(0, 10, 2):
    print(i)
```

```python
# print even numbers from 1 to 20
for number in range(1,21):
    if number%2==0:
        print(number)
```

```python
# when to use for vs while loops
# for loops: when the number of iterations is known
# while loops: when the number of iterations is not known and depends on a condition
count= 0
while count<5:
    print("Count is:", count)
    count+=1
    
```

```python
# building a secret guessing game
secret_number= 7
user_guess= int(input("Guess the secret number (between 1 and 10): "))
if user_guess== secret_number:
    print("Congratulations! You guessed the secret number.")
else:
    print("Sorry, wrong guess. Try again!")
```
